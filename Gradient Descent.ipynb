The Gradient descent intercept,Slope = 102.04225167807915 ,  0.013982492881876953

The Sklearn intercept,slope = 102.83201911 , 0.00318316

```python
import numpy as np
import matplotlib.pyplot as plt
import time
```

## Random Gen


```python
x = np.random.randint(1,200,1000)
y = np.random.randint(1,200,1000)
```

## Gradient

```python
import numpy as np

slope = 0
intercept = 0
learning= 0.00001

epochs = 1000000

for i in range(epochs):
    y_pred = intercept + slope*x
    intercept -= (learning*2*(y_pred - y).mean())
    slope -= (learning*2*(x*(y_pred - y)).mean())
    
print(intercept,slope)
```

## Sklearn

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

x = x.reshape((len(x),1))
y = y.reshape((len(x),1))
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.025)
regr = LinearRegression()

regr.fit(X_train, y_train)
```
