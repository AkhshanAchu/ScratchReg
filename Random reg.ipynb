```python
x = np.random.randint(1,200,1000)
y = np.random.randint(1,200,1000)
```

```python
xtest,xtrain = x[:100],x[100:]
ytest,ytrain = y[:100],y[100:]
```

```python
ytrain.T.shape
```

```python
xtrain.shape
```

```python
ytrain.shape
```

```python
intercept = ((sum(ytrain)*(xtrain@xtrain.T) - (sum(xtrain)*(xtrain@ytrain.T)))/((xtrain@xtrain.T)-sum(xtrain)**2))
```

```python
intercept
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(xtrain,ytrain)
```

```python
slope = (len(xtrain)*((xtrain@ytrain.T))-(sum(xtrain)*sum(ytrain)))/((len(xtrain)*(xtrain@xtrain.T))-sum(xtrain)**2)
```

```python
slope
```

```python
intercept = np.mean(ytrain) - slope*np.mean(xtrain)
```

```python
intercept
```

```python
w=[intercept,slope]
```

```python
xtest.shape
```

```python
xtest = xtest.reshape((xtest.shape[0],1))
```

```python
xtest.shape
```

```python
xtest = np.append(np.ones((len(xtest),1)),xtest,axis=1)
```

```python
ypred = xtest@w
```

```python
ypred
```

```python
plt.scatter( xtest[:,1], ytest, color = 'blue' )
      
plt.plot( xtest[:,1], ypred , color = 'orange' )
```
